# Problem 2 #
**write a python Program to arithmetical operations addition and division.**

```python
# Addition
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
sum_result = num1 + num2
print(f"Sum: {num1} + {num2} = {sum_result}")
```

```python
#Division
num3 = float(input("Enter the first number: "))
num4 = float(input("Enter the second number: "))
if num4 == 0:
    print("Error: Division by Zero is not allowed.")
else:
    div_result = num3 / num4
    print(f"Division: {num3} / {num4} = {div_result}")
```

# Problem 3 #
**write a python Program to find the area of a triangle.**

```python
base = float(input("Enter the length of the base of the triangle: "))
height = float(input("Enter the length of the height of the triangle: "))
area = 1/2 * base * height
print(f"The area of the triangle is : {area}")
```
